# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

 Seaborn is a popular Python data visualization library that provides a high-level interface for creating informative and attractive statistical graphics. Here are five types of plots you can create using Seaborn, along with their common uses:

Scatter Plot:

Use: Scatter plots are used to visualize the relationship between two continuous variables. Each point on the plot represents an observation in the dataset, with one variable plotted on the x-axis and the other on the y-axis.
Common Uses: Identifying patterns, clusters, or trends in data, and assessing the correlation between two variables.
Line Plot:

Use: Line plots display data points connected by lines. They are typically used to visualize trends over time or some ordered category.
Common Uses: Showing changes in data over a continuous range, such as time-series data or trends in experimental data.
Bar Plot:

Use: Bar plots represent categorical data with rectangular bars. One axis (usually the x-axis) represents the categories, while the other axis represents the values associated with those categories.
Common Uses: Comparing quantities across different categories, displaying frequency distributions, and showing how a single categorical variable is distributed.
Histogram:

Use: Histograms are used to visualize the distribution of a single variable. They divide the data into bins and display the frequency or density of observations within each bin.
Common Uses: Understanding the underlying distribution of a continuous variable, identifying modes, skewness, and presence of outliers.
Heatmap:

Use: Heatmaps display data in a matrix format using colors to represent the values. They are particularly useful for visualizing relationships in large datasets by displaying values as colors.
Common Uses: Displaying correlations between multiple variables, visualizing matrices (e.g., confusion matrices), and showing patterns in two-dimensional datasets.

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions. 

# Note:  timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset('titanic')
titanic
```

```python
# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Create the first box plot: age vs. pclass
sns.boxplot(ax=axes[0], data=titanic, x='pclass', y='age')
axes[0].set_title('Box Plot of Age vs. Pclass')
axes[0].set_xlabel('Pclass')
axes[0].set_ylabel('Age')

# Create the second box plot: fare vs. pclass
sns.boxplot(ax=axes[1], data=titanic, x='pclass', y='fare')
axes[1].set_title('Box Plot of Fare vs. Pclass')
axes[1].set_xlabel('Pclass')
axes[1].set_ylabel('Fare')

# Adjust layout
plt.tight_layout()
plt.show()
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset. 

```python
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
diamonds
```

```python
sns.histplot(data = diamonds , x=diamonds.price, y=diamonds.cut, hue = diamonds.cut)
```

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
flights = sns.load_dataset('flights')
flights
```

```python
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(data=flights_pivot, cmap='YlGnBu', annot=True, fmt='d')
```
